```python
def data_read(file_name):
    open_file =open(file_name,encoding="utf8")
    from csv import reader
    read_file = reader(open_file)
    return list(read_file)
apple_data = data_read('AppleStore.csv')
google_data = data_read('googleplaystore.csv')
```

```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
```

```python
review_max = {}
for row in google_data[1:] :
    app_name = row[0]
    n_reviews = float(row[3])
    if app_name  in review_max and review_max[app_name] < n_reviews :
        review_max[app_name] = n_reviews
    if app_name not in review_max:
        review_max[app_name] = n_reviews
print(len(review_max))

        
    
    
```

```python
android_clean = []
already_added = []
for row in google_data[1:] :
    name =  row[0]
    n_reviews = float(row[3])
    if n_reviews == review_max[name] and name not in already_added :
        android_clean.append(row)
        already_added.append(name)
len(android_clean)
print(android_clean[0:5])
```

```python
def str_numb(strng):
    number_of_characters = 0
    for character in strng :
        if ord(character) > 127 :
            number_of_characters+=1
    if number_of_characters > 3 :
        return False
    else :
        return True
        

print(str_numb('Instagram'))     
print(str_numb('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))  
print(str_numb('Docs To Go‚Ñ¢ Free Office Suite')) 
print(str_numb('Instachat üòú')) 
```

```python
def freq_table(dataset,indx):
    temp_dict ={}
    for row in dataset :
        val = row[indx]
        if val in temp_dict :
            temp_dict[val] += 1
        else :
            temp_dict[val] = 1
    for key in temp_dict:
        temp_dict[key]=(temp_dict[key]/len(dataset))*100
    return temp_dict

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])


```

```python
display_table(apple_eng_free_apps,11)    
```

```python
display_table(google_eng_free_apps,1) #google_category_freq   
```

```python
 display_table(google_eng_free_apps,9) #google_genre_freq 
```

```python
apple_genre_freq = freq_table(apple_eng_free_apps,11)
for genre in apple_genre_freq :
    total = 0
    len_genre = 0
    for row in apple_eng_free_apps:
        genre_app =row[11]
        if genre_app == genre :
            total += float(row[5])
            len_genre += 1
    avg_user_rating = total/len_genre
    print(genre,":",avg_user_rating) 
```

```python
google_category_freq = freq_table(google_eng_free_apps,1)
for category in google_category_freq :
    total = 0
    len_category = 0
    for row in google_eng_free_apps :
        category_app = row[1]
        if category_app == category :
            installs = row[5]
            n_installs = float(installs.replace('+','').replace(',',''))
            total += n_installs
            len_category += 1
    avg_user_rating = total/len_category
    
            
```
